# Data Cleaning

now we clean the data set by filing the missing values with the mean of its column. As all the columns having missing values have no non-numeric entry, so we use mean to fill the missing values. But if it had non-negative enries we would have used mode to fill the specific columns

```python
#fill years resale value column
df['__year_resale_value'].fillna(df['__year_resale_value'].mean(),inplace=True)
```

```python
df['__year_resale_value'].isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
sns.heatmap(df.isna())
```

we can see that there are no null values in the dataframe

```python
#seperate dependent and independent variable
X=df[['Sales_in_thousands','__year_resale_value','Engine_size', 'Horsepower',
       'Wheelbase', 'Width', 'Length', 'Curb_weight','Fuel_capacity','Fuel_efficiency','Power_perf_factor']]
y=df['Price_in_thousands']
```

Here **y is independent variable** and **X is dependent variable class**

# Data Modelling


**Building model using Linear Regression**

```python
#import train test split methord
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#importing Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
#fit the Linear Regression
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
#now we see the cofficient of different varibles
Coeff_df= pd.DataFrame(lm.coef_,X.columns,columns=['coefficient'])
```

```python
Coeff_df
```

**This depicts, how 1 unit increase in independent variable (price) leads to changes in different dependent variable**

```python
#predicting the model
predict=lm.predict(X_test)
```

# Testing

**see how the model is performing**

```python
#plotting the scatter plot
plt.scatter(y_test,predict)
```

```python
#calculating the R squared values
from sklearn.metrics import r2_score
r2_score(y_test,predict)
```

```python
#getting the distplot of the same
sns.distplot((y_test-predict),bins=50)
```

**For a model to be a good fit Rsquared value should be more than .91. But here the R squared value is exactly 1 and the scatter plot diagram is perfectly linear which is not feasible. Therefore the moddel is overfitting. This can happen due to the discripencies in the dataset.**
